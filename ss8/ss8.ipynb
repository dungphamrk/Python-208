```python
# Bài 5
import pandas as pd

df_students = pd.read_csv('students.csv')
df_scores = pd.read_csv('scores.csv')

merged = pd.merge(df_students, df_scores, on='id')

report = merged.groupby(['id', 'name'])['score'].agg(['min', 'max', 'mean']).reset_index()

report.to_csv('report.csv', index=False)

print(report)
```

```python
# Bài 6
import numpy as np

a = np.array([[1],[2],[3],[4]])
b = np.array([[10,20,30,40,50]])

c = a * b
print("a*b =")
print(c)

print("\nsqrt =")
print(np.sqrt(c))

print("\nlog =")
print(np.log(c))

print("\nmin theo axis=1 =", np.min(c, axis=1))
print("max theo axis=1 =", np.max(c, axis=1))
print("mean theo axis=1 =", np.mean(c, axis=1))
print("std theo axis=1 =", np.std(c, axis=1))
```

```python
# Bài 7
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "name": ["An", "Bình", "Châu", "Dũng"],
    "age": [20, np.nan, 22, 21],
    "score": [8.5, np.nan, 7.0, 9.0],
    "city": ["Hà Nội", "HCM", np.nan, "Hà Nội"]
})

df["age"] = df["age"].fillna(df["age"].mean())
df["score"] = df["score"].fillna(df["score"].median())
df["city"] = df["city"].fillna("Unknown")
df["city"] = df["city"].astype("category")

def score_level(x):
    if x >= 8: return "Good"
    elif x >= 7: return "Average"
    return "Bad"

df["score_level"] = df["score"].apply(score_level)

age_mean = df["age"].mean()
age_std = df["age"].std()
df["age_normalized"] = (df["age"] - age_mean) / age_std
df
```

```python
# Bài 8
import pandas as pd
from pathlib import Path
import json
from datetime import datetime

log_dir = Path("logs/")
data = []

for file_path in log_dir.glob("*.json"):
    date_str = file_path.stem.split("_")[1]
    date = datetime.strptime(date_str, "%Y-%m-%d").date()
    
    # Đọc nội dung JSON
    with open(file_path, "r", encoding="utf-8") as f:
        log_entry = json.load(f)
        log_entry["date"] = date 
        data.append(log_entry)

df = pd.DataFrame(data)

daily_report = df.groupby("date")["duration"].sum().reset_index()

daily_report.to_csv("daily_report.csv", index=False)

print("Đã xuất báo cáo daily_report.csv thành công!")
```
