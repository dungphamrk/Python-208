```python
# Bt8
products = {
    "A": 120,
    "B": 80,
    "C": 300,
    "D": 50
}
if "A" in products:
    print("Gi√° s·∫£n ph·∫©m A:", products["A"])
else:
    print("S·∫£n ph·∫©m A ƒë√£ b·ªã x√≥a")

sorted_products = dict(sorted(products.items(), key=lambda item: item[1]))
print("Danh s√°ch s·∫£n ph·∫©m sau khi s·∫Øp x·∫øp theo gi√°:", sorted_products)

total_price = sum(products.values())
print("T·ªïng gi√° tr·ªã c·ªßa t·∫•t c·∫£ s·∫£n ph·∫©m:", total_price)

products = {k: v for k, v in products.items() if v <= 200}
print("Danh s√°ch s·∫£n ph·∫©m sau khi x√≥a gi√° > 200:", products)
```

```python

listProduct = []


def them_san_pham_moi(id=id):
    name = input("Nh·∫≠p t√™n s·∫£n ph·∫©m: ")
    quantity = input("Nh·∫≠p s·ªë l∆∞·ª£ng s·∫£n ph·∫©m: ")
    price = input("Nh·∫≠p gi√° s·∫£n ph·∫©m: ")
    newProduct = (id, name, quantity, price)
    listProduct.append(newProduct)


def hien_thi_danh_sach_san_pham():
    for product in listProduct:
        print(f"name={product[1]},quantity={product[2]},price={product[3]}")


def tim_kiem_san_pham_theo_ma():
    id = int(input("Nh·∫≠p m√£ s·∫£n ph·∫©m c·∫ßn t√¨m: "))
    for product in listProduct:
        if product[0] == id:
            print(f"name={product[1]},quantity={product[2]},price={product[3]}")
        else:
            print("Kh√¥ng t√¨m th·∫•y s·∫£n ph·∫©m v·ªõi m√£ ƒë√£ nh·∫≠p.")


def cap_nhat_thong_tin_san_pham():
    id = int(input("Nh·∫≠p m√£ s·∫£n ph·∫©m c·∫ßn c·∫≠p nh·∫≠t: "))
    for product in listProduct:
        if product[0] == id:
            name = input("Nh·∫≠p t√™n s·∫£n ph·∫©m m·ªõi: ")
            quantity = input("Nh·∫≠p s·ªë l∆∞·ª£ng s·∫£n ph·∫©m m·ªõi: ")
            price = input("Nh·∫≠p gi√° s·∫£n ph·∫©m m·ªõi: ")
            updatedProduct = (id, name, quantity, price)
            listProduct.remove(product)
            listProduct.append(updatedProduct)
            print("C·∫≠p nh·∫≠t s·∫£n ph·∫©m th√†nh c√¥ng.")
            return


def xoa_san_pham_theo_ma():
    id = int(input("Nh·∫≠p m√£ s·∫£n ph·∫©m c·∫ßn x√≥a: "))
    for product in listProduct:
        if product[0] == id:
            listProduct.remove(product)
            print("X√≥a s·∫£n ph·∫©m th√†nh c√¥ng.")
            return


def hien_thi_menu():
    print("\n" + "="*30)
    print("      üìú MENU QU·∫¢N L√ù S·∫¢N PH·∫®M")
    print("="*30)
    print("1. Th√™m s·∫£n ph·∫©m m·ªõi.")
    print("2. Hi·ªÉn th·ªã danh s√°ch s·∫£n ph·∫©m.")
    print("3. T√¨m ki·∫øm s·∫£n ph·∫©m theo m√£ s·∫£n ph·∫©m.")
    print("4. C·∫≠p nh·∫≠t th√¥ng tin s·∫£n ph·∫©m.")
    print("5. X√≥a s·∫£n ph·∫©m theo m√£ s·∫£n ph·∫©m.")
    print("0. Tho√°t ch∆∞∆°ng tr√¨nh.")
    print("-" * 30)


def main():
    while True:
        hien_thi_menu()
        id = 1
        lua_chon = input("‚ñ∂Ô∏è Vui l√≤ng nh·∫≠p l·ª±a ch·ªçn c·ªßa b·∫°n (0-5): ")

        if lua_chon == '1':
            them_san_pham_moi(id)
        elif lua_chon == '2':
            hien_thi_danh_sach_san_pham()
        elif lua_chon == '3':
            tim_kiem_san_pham_theo_ma()
        elif lua_chon == '4':
            cap_nhat_thong_tin_san_pham()
        elif lua_chon == '5':
            xoa_san_pham_theo_ma()
        elif lua_chon == '0':
            print("\nüëã ƒê√£ tho√°t ch∆∞∆°ng tr√¨nh. T·∫°m bi·ªát!")
            break
        else:
            print("\n‚ùå L·ª±a ch·ªçn kh√¥ng h·ª£p l·ªá. Vui l√≤ng nh·∫≠p l·∫°i (0-5).")
        id += 1


if __name__ == "__main__":
    main()
```
