```python
# Bt1
import numpy as np

arr1 = np.arange(1, 11)
print("Mảng 1 chiều:", arr1)

print("Shape:", arr1.shape)
print("Dtype:", arr1.dtype)

arr1_float = arr1.astype(float)
print("Mảng float:", arr1_float)
print("Dtype mới:", arr1_float.dtype)


arr2 = np.arange(1, 10).reshape(3, 3)
print("\nMảng 2 chiều:")
print(arr2)

print("Shape:", arr2.shape)
print("Dtype:", arr2.dtype)

arr2_float = arr2.astype(float)
print("Mảng 2D float:")
print(arr2_float)
print("Dtype mới:", arr2_float.dtype)

 
```

```python
# Bài 3
import numpy as np

arr = np.array([[88, 35, 62], [11, 98, 34], [64, 24, 48]])

print(f"min: {arr.min()}")
print(f"max: {arr.max()}")
print(f"mean: {arr.mean()}")
print(f"sum: {arr.sum()}")
print(f"std: {arr.std()}")

print(f"sum of column: {arr.sum(axis=0)}")
print(f"mean of column: {arr.mean(axis=0)}")

print(f"sum of row: {arr.sum(axis=1)}")
print(f"mean of row: {arr.mean(axis=1)}")
```

```python
# Bài 5
import numpy as np
import time

arr = np.random.randint(1, 21, 10)
squared = np.multiply(arr, arr)
root = np.sqrt(arr)

added = np.add(arr, 5)
multiplied = np.multiply(arr, 3)
root_ufunc = np.sqrt(arr)

arr2d = np.random.randint(1, 100, (1000, 1000))

start_vec = time.time()
vec_result = arr2d * 2 + 3
end_vec = time.time()

start_loop = time.time()
loop_result = np.zeros_like(arr2d)
for i in range(arr2d.shape[0]):
    for j in range(arr2d.shape[1]):
        loop_result[i, j] = arr2d[i, j] * 2 + 3
end_loop = time.time()

print("1D Array:", arr)
print("squared:", squared)
print("squareRoot:", root)
print("add:", added)
print("multiply:", multiplied)
print("sqrtUfunc:", root_ufunc)
print("vectorized time:", end_vec - start_vec)
print("loop time:", end_loop - start_loop)
```

```python
# Bài 7
import numpy as np
from datetime import datetime

start = datetime.now()
arr1 = np.array([[ 8, 12, 14,  6,  4],[18, 13,  2,  2, 12],[ 6, 18,  5, 10, 10],[ 9,  2,  4, 12, 12],[ 2,  3,  2, 17,  3]])
arr2 = np.array([10, 20, 30, 40, 50])

print(f"total: {arr1 + arr2}")
print(f"diff: {arr1 - arr2}")
print(f"root: {arr1**(1/2)}")
print(f"sum 5: {arr1 + 5}")

end = datetime.now()
print(f"Time to done: {end - start}")
```

```python
# Bài 8
import numpy as np


maxtrix = np.array([[6,9,7],[2,1,5],[7,8,3]])

if np.linalg.det(maxtrix) != 0:
    maxtrix_inv = np.linalg.inv(maxtrix)
    print("\nInverse matrix:\n", maxtrix_inv)
else:
    print("Matrix is not invertible.")

eig_values, eig_vectors = np.linalg.eig(maxtrix)

print("\nEigenvalues:", eig_values)
print("\nEigenvectors:\n", eig_vectors)
```
